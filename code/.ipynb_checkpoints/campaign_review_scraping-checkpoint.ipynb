```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ 1. ë“œë¼ì´ë²„ ì„¤ì • â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
options = webdriver.ChromeOptions()
# options.add_argument('--headless')  # ì°½ ì•ˆ ë„ìš°ê³  ì‹¶ìœ¼ë©´ ì£¼ì„ í•´ì œ
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
wait = WebDriverWait(driver, 10)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ 2. ë°˜ë³µ êµ¬ê°„ ì´ˆê¸°í™” â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
start = 1
end = 101
file_no = 1  # ì €ì¥ìš© ë²ˆí˜¸

while True:
    comments_all = []

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ 3. í˜ì´ì§€ ë°˜ë³µ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    for page in range(start, end):
        URL = f"https://www.momq.co.kr/shop/neoevent_view.html?ev_no=202004180005&cmt_type=&page={page}"
        try:
            driver.get(URL)

            # ëŒ“ê¸€ ì˜ì—­ ëŒ€ê¸°
            ul = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="commentWrap"]/div[3]/ul')))
            driver.execute_script("arguments[0].scrollIntoView(false);", ul)
            time.sleep(0.4)

            # íŒŒì‹±
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            lis = soup.select('#commentWrap > div.commentListArea > ul > li')
            if not lis:
                print(f"âŒ {page}page: ëŒ“ê¸€ ì—†ìŒ")
                continue

            for li in lis:
                cmt = li.select_one('div > p')
                dt = li.select_one('div > div > span')
                comments_all.append({
                    "page": page,
                    "content": cmt.get_text(strip=True) if cmt else '',
                    "date": dt.get_text(strip=True) if dt else ''
                })

            print(f"âœ… {page}page â†’ {len(lis)}ê°œ ìˆ˜ì§‘")

        except TimeoutException:
            print(f"âš ï¸ {page}page: ë¡œë”© ì‹¤íŒ¨")

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ 4. ì €ì¥ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    df = pd.DataFrame(comments_all)
    df.to_csv(f"./review/momq_comments_{file_no}.csv", index=False, encoding="utf-8-sig")
    print(f"\nğŸ‰ ì´ {len(df)}ê°œ ì €ì¥ ì™„ë£Œ â†’ momq_comments_{file_no}.csv")

    # ë‹¤ìŒ êµ¬ê°„ ì„¤ì •
    start += 100
    end += 100
    file_no += 1

    if end > 2600:
        break

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ 5. ë“œë¼ì´ë²„ ì¢…ë£Œ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
driver.quit()
```
