```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1)  & CohortMonth 
df['OrderMonth'] = df[''].dt.to_period('M')
first_month = df.groupby('')['OrderMonth'].min().rename('CohortMonth')
df = df.join(first_month, on='')

# 2) CohortIndex ( ) 
df['CohortIndex'] = (df['OrderMonth'].dt.year - df['CohortMonth'].dt.year) * 12 + \
                    (df['OrderMonth'].dt.month - df['CohortMonth'].dt.month) + 1

# 3) Cohort   
cohort_data = df.groupby(['CohortMonth', 'CohortIndex'])[''] \
                .nunique().reset_index()

# 4) Pivot &  
cohort_pivot = cohort_data.pivot(index='CohortMonth', columns='CohortIndex', values='')
retention = cohort_pivot.divide(cohort_pivot.iloc[:, 0], axis=0)

# 5) Heatmap 
plt.figure(figsize=(10, 8))

mask = np.zeros_like(retention, dtype=bool)
mask[:, 0] = True

sns.heatmap(
    retention, 
    annot=True, fmt=".0%", cmap="YlGnBu",
    mask=mask
)

plt.title("Cohort Retention Rate")
plt.ylabel("Cohort Month")
plt.xlabel("Months Since First Purchase")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'

# 1.    (pivot)
ratio_df = (
    df.groupby('')['']
    .value_counts(normalize=True)
    .rename('')
    .reset_index()
    .pivot(index='', columns='', values='')
    .fillna(0)
)

# 2. 
plt.figure(figsize=(10, 6))
bottom = pd.Series([0] * len(ratio_df), index=ratio_df.index)
colors = plt.cm.tab20.colors  # 20   

for i, column in enumerate(ratio_df.columns):
    plt.bar(ratio_df.index, ratio_df[column], bottom=bottom, label=column, color=colors[i % len(colors)])
    
    #   
    for idx in ratio_df.index:
        val = ratio_df.loc[idx, column]
        if val > 0.03:
            plt.text(idx, bottom[idx] + val / 2, f"{val:.0%}", ha='center', va='center', fontsize=13)

    bottom += ratio_df[column]

# 3.  
plt.title('   ()', fontsize=13)
plt.ylabel('')
plt.xlabel('')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.legend(title='', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#  <->  
print('  <->  ')
tab = pd.crosstab(df[''], df[''])

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(tab, correction=True)
print(f" = {chi2:.2f}")
print(f"p-value = {p:.3f}")
print(f"degrees of freedom = {dof}")

#    
expected_df = pd.DataFrame(expected, index=tab.index, columns=tab.columns)
print("\nExpected frequencies:")
print(expected_df)
```
